Changing data format in "Trip Miles" from x.xxx,xx
to xxxx.xx and then converting to float

```python
data["Trip Miles"] = data["Trip Miles"].str.replace(".", "", regex=False)
data["Trip Miles"] = data["Trip Miles"].str.replace(",", ".", regex=False)
data["Trip Miles"] = pd.to_numeric(data["Trip Miles"])
```

```python
#print(data["Trip Miles"])
print(data["Trip Miles"].mean())
print(data["Trip Miles"].max())
print(data["Trip Miles"].min())
```

What are the differences between Extras and Tips
How do Tolls and Fares look like

Change Extras, Tips, Tolls, Fare to numerical 

```python
data["Extras"] = data["Extras"].str.replace(".", "", regex=False)
data["Extras"] = data["Extras"].str.replace(",", ".", regex=False)
data["Extras"] = data["Extras"].str.replace("$", "", regex=False)
data["Extras"] = pd.to_numeric(data["Extras"])
data["Tips"] = data["Tips"].str.replace(".", "", regex=False)
data["Tips"] = data["Tips"].str.replace(",", ".", regex=False)
data["Tips"] = data["Tips"].str.replace("$", "", regex=False)
data["Tips"] = pd.to_numeric(data["Tips"])
data["Tolls"] = data["Tolls"].str.replace(".", "", regex=False)
data["Tolls"] = data["Tolls"].str.replace(",", ".", regex=False)
data["Tolls"] = data["Tolls"].str.replace("$", "", regex=False)
data["Tolls"] = pd.to_numeric(data["Tolls"])
data["Fare"] = data["Fare"].str.replace(".", "", regex=False)
data["Fare"] = data["Fare"].str.replace(",", ".", regex=False)
data["Fare"] = data["Fare"].str.replace("$", "", regex=False)
data["Fare"] = pd.to_numeric(data["Fare"])
```

```python
data["Trip Total"] = data["Trip Total"].str.replace(".", "", regex=False)
data["Trip Total"] = data["Trip Total"].str.replace(",", ".", regex=False)
data["Trip Total"] = data["Trip Total"].str.replace("$", "", regex=False)
data["Trip Total"] = pd.to_numeric(data["Trip Total"])
```
